# **Scatter plot**

A scatter plot is a two-dimensional graph for visualizing data that uses points to represent the value between two different variables.

One of the variables corresponds to the ùë• value and the other variable corresponds to the ùë¶ value.

This type of graph is ideal when you want to show the relationship between two variables, and it gives great indications about the existing correlation between them.

unnamed-chunk-1-1.png


It's worth emphasize that not all connections are linear. Many of them follow other types of non-linear patterns or do not follow any apparent pattern at all.

# **Tips**

Use to show connection between two variables

* Allows you to visually identify if there is any pattern
* Helps to reduce the dimensionality of the problem
* Very common in Data Science

# **Problems**

* It's not used very often in the business world
* People confuse correlation with causation

# **Plot the Scatter Plot**

```python
# import the necessary packages
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/housing.csv"

# import the dataset
df = pd.read_csv(DATA_PATH)

# see the first entries
df.head()
```

```python
# plot the scatter plot
fig, ax = plt.subplots()

# plot median_income vs. median_house_value
df.plot.scatter(x='median_income', y='median_house_value', ax=ax, s=.5)

plt.show()
```

# Linear Regression on a Scatter Plot

Just for the sake of curiosity and knowledge, in a distribution graph it is possible to find a line that best represents all those plotted points. See how to plot a line using Linear Regression using the seaborn library and the sns.regplot() method.

```python
# import Seaborn
import seaborn as sns


# plot scatter plot and linear regression
fig, ax = plt.subplots()
ax.set_xlim(right=8)
sns.regplot('median_income', 'median_house_value', scatter_kws={"s": 0.2},
            line_kws={"color": "red"}, data=df, ax=ax)

plt.show()
```
